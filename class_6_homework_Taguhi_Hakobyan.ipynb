```python
import requests
from bs4 import BeautifulSoup
```

```python
my_list=['teacher', 'lawyer', 'data-scientist','economist','doctor','waitress','journalist','actor','musician','CEO']
```

```python
def indeed_url_maker(job):
    """makes an url of the job you are interested"""
    ### put your code here
    return 'https://www.indeed.com/q-'+job+'-jobs'
```

```python
my_list_url=[]
for job in my_list:
    print(job)
    my_list_url.append(indeed_url_maker(job))
    print(indeed_url_maker(job))
    my_list_url.append(indeed_url_maker(job))
print(my_list_url)
```

```python
res = requests.get("https://www.indeed.com/q-teacher-jobs")
soup = BeautifulSoup(res.content, 'html.parser')
```

```python
relevant_companies={}
for job in my_list:
    relevant_companies[job]=[]
    url=indeed_url_maker(job)
 ###    print(url)
    res = requests.get(url)
    soup = BeautifulSoup(res.content, 'html.parser')
    for i in range(len(soup.find_all('span', class_="company"))):
        relevant_companies[job].append(soup.find_all('span', class_="company")[i].text)
```
